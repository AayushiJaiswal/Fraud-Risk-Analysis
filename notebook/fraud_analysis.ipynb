```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("creditcard.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Class'].value_counts()
```

```python
sns.countplot(x='Class',data=df)
plt.title("Fraud vs Normal Transactions")
plt.savefig("Fraud_vs_Normal.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Class',y='Amount',data=df)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Time'],bins=50)
plt.savefig("time_histogram.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Class Imbalance Analysis
fraud_count=df['Class'].value_counts()
fraud_percentage=df['Class'].value_counts(normalize=True) * 100
print(fraud_count)
print(fraud_percentage)
```

```python
#Transaction Amount Analysis
df.groupby('Class')['Amount'].describe()
plt.figure(figsize=(8,5))
sns.boxplot(x='Class',y='Amount',data= df)
plt.yscale('log')
plt.title("Transaction Amount Distribution")
plt.savefig("time_amount_boxplot.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Correlation with Fraud
correlation=df.corr()['Class'].sort_values(ascending=False)
print(correlation.head(10))
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.savefig("corr_heatmap.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Time Based Fraud Pattern
df['Hour'] = df['Time'] // 3600 % 24
plt.figure(figsize=(10,5))
sns.countplot(x='Hour', hue='Class', data=df)
plt.title("Fraud Distribution by Hour")
plt.savefig("time_based_analysis.png", dpi=300, bbox_inches='tight')
plt.show()
```
